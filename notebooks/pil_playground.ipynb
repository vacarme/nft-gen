```python
from random import randint
```

```python
img = Image.new('RGB', size=(255,255), color=(0,0,0))
img
```

```python
image_size = 255
```

```python
draw = ImageDraw.Draw(img)
```

```python
points = [(randint(0,255), randint(0,255)) for _ in range(10)]
points
```

```python
min_x = min([p[0] for p in points])
max_x = max([p[0] for p in points])
min_y = min([p[1] for p in points])
max_y = max([p[1] for p in points])
```

```python
delta_x = min_x - (image_size - max_x)
delta_y = min_y - (image_size - max_y)
points = [(point[0] - delta_x // 2, point[1] - delta_y // 2) for point in points]
```

```python
min_x = min([p[0] for p in points])
max_x = max([p[0] for p in points])
min_y = min([p[1] for p in points])
max_y = max([p[1] for p in points])
```

```python
draw.rectangle((min_x, min_y, max_x, max_y), outline=(255,0,0))
```

```python
lines = [(points[i], points[i+1]) for i in range(len(points)-1)]
```

```python
lines[0]
```

```python
for line in lines:
    draw = ImageDraw.Draw(img)
    draw.line(line, fill=line[0]+(randint(0,255),))
```
